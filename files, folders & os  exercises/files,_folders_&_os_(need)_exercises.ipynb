```python
import os

def rename_my_file(old_file_name):
    # Define the dictionary mapping month abbreviations to numerical format
    month_mapping = {
        'jan': '01', 'feb': '02', 'mar': '03',
        'apr': '04', 'may': '05', 'jun': '06',
        'jul': '07', 'aug': '08', 'sep': '09',
        'oct': '10', 'nov': '11', 'dec': '12'
    }

    # Split the filename into directory, name, and extension
    directory, file_name = os.path.split(old_file_name)
    name, extension = os.path.splitext(file_name)

    # Extract month, day, and city from the filename
    month, day_city = name.split('-')
    day, city = day_city.split('_')

    # Convert month abbreviation to numerical format
    month_numeric = month_mapping[month]

    # Create the new filename with the desired format
    new_filename = os.path.join(directory, f"{month_numeric}-{day}_{city}{extension}")

    return new_filename

# Test the function
old_file_name = 'os-collaboration-exercise-data\\oct-08_singapore.txt'
new_file_name = rename_my_file(old_file_name)
print(new_file_name)
```

**Task 7**

```python
import os
import glob
import shutil

# List of cities
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
          'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

# Iterate over each city
for city in cities:
    # List all files related to the current city using glob
    city_files = glob.glob(f'os-collaboration-exercise-data/**/*_{city.lower()}*', recursive=True)
    
    # Iterate over each file and move it to the corresponding city folder
    for file_path in city_files:
        # Get the filename from the file path
        file_name = os.path.basename(file_path)
        
        # Construct the destination path in the city folder
        destination_path = os.path.join(city, file_name)
        
        # Move the file to the city folder
        shutil.move(file_path, destination_path)
        
        print(f"Moved {file_name} to {city} folder.")
```
