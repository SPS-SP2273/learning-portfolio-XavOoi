**Exercise 1**

```python
from matplotlib import pyplot as plt
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr=x_err, yerr=y_err,
             color="red", linestyle="dashed",
             marker="o", markerfacecolor="yellow",
             ecolor="black",capsize=5,
             label="Y values with errors")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc="lower right")
plt.show()
```

**Exercise 2**

```python
import numpy as np
```

```python
x = np.linspace(0, 50, 100)
y=np.exp(-x/10)*np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('y=e^{-x/10}*sin(x)')
plt.grid(True, alpha=.25) # makes a grid and sets it to 25% opacity
plt.show()
```

```python
x = np.linspace(0, 50, 100)
y=np.exp(-x/10)*np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,1) # Set limits of y axis to -1 to 1
plt.title('y=e^{-x/10}*sin(x)')
plt.grid(True, alpha=.25) # makes a grid and sets it to 25% opacity
plt.show()
```

```python
x = np.linspace(0, 50, 100)
y=np.exp(-x/10)*np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1,1) # Set limits of y axis to -1 to 1
plt.title(r'$y=e^{-x/10}sin(x)$', size=20) # to make an equation for the title; font size 20
plt.grid(True, alpha=.25) # makes a grid and sets it to 25% opacity
plt.show()
```

```python
x = np.linspace(0, 50, 100)
y=np.exp(-x/10)*np.sin(x)

plt.plot(x, y)
plt.xlabel('x', c='darkorange', size=15) # set x axis as 'x' font size 15 and in darkorange colour
plt.ylabel('y', c='darkorange', size=15, rotation=0) # set y axis as 'y' font size 15 and in darkorange colour and 
                                                     # rotate to upright position
plt.ylim(-1,1) # Set limits of y axis to -1 to 1
plt.title(r'$y=e^{-x/10}sin(x)$', size=20) # to make an equation for the title
plt.grid(True, alpha=.25) # makes a grid and sets it to 25% opacity
plt.show()
```
